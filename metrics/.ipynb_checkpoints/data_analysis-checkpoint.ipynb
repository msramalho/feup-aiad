```python
# Reading Data
Location = '.\merged_all.csv'
df = pd.read_csv(Location)
```

```python
# Check data type of the columns
df.dtypes
```

```python
# Reparting data from different experiences with different sizes
lab_10_10 = df.loc[df['xsize'] == 10].loc[df['ysize'] == 10]
lab_50_20 = df.loc[df['xsize'] == 50].loc[df['ysize'] == 20]
lab_20_50 = df.loc[df['xsize'] == 20].loc[df['ysize'] == 50]
lab_60_40 = df.loc[df['xsize'] == 60].loc[df['ysize'] == 40]
lab_40_60 = df.loc[df['xsize'] == 40].loc[df['ysize'] == 60]
lab_100_100 = df.loc[df['xsize'] == 100].loc[df['ysize'] == 100]
```

```python
# Getting average of averageSpeed lab_10_10
average_10_10_value = lab_10_10['averageSpeed'].mean()
print('Average in global of 10x10: ' + str(average_10_10_value) + ' in a total of ' + str(lab_10_10.size))
print('')

# Average of different types agents
lab_10_10_BacktrackAgent = lab_10_10.loc[lab_10_10['agentType'] == 'BacktrackAgent']
lab_50_20_BacktrackAgent = lab_50_20.loc[lab_50_20['agentType'] == 'BacktrackAgent']
lab_20_50_BacktrackAgent = lab_20_50.loc[lab_20_50['agentType'] == 'BacktrackAgent']
lab_60_40_BacktrackAgent = lab_60_40.loc[lab_60_40['agentType'] == 'BacktrackAgent']
lab_40_60_BacktrackAgent = lab_40_60.loc[lab_40_60['agentType'] == 'BacktrackAgent']
lab_100_100_BacktrackAgent = lab_100_100.loc[lab_100_100['agentType'] == 'BacktrackAgent']

lab_10_10_ForwardAgent = lab_10_10.loc[lab_10_10['agentType'] == 'ForwardAgent']
lab_50_20_ForwardAgent = lab_50_20.loc[lab_50_20['agentType'] == 'ForwardAgent']
lab_20_50_ForwardAgent = lab_20_50.loc[lab_20_50['agentType'] == 'ForwardAgent']
lab_60_40_ForwardAgent = lab_60_40.loc[lab_60_40['agentType'] == 'ForwardAgent']
lab_40_60_ForwardAgent = lab_40_60.loc[lab_40_60['agentType'] == 'ForwardAgent']
lab_100_100_ForwardAgent = lab_100_100.loc[lab_100_100['agentType'] == 'ForwardAgent']

lab_10_10_NegotiatingAgent = lab_10_10.loc[lab_10_10['agentType'] == 'NegotiatingAgent']
lab_50_20_NegotiatingAgent = lab_50_20.loc[lab_50_20['agentType'] == 'NegotiatingAgent']
lab_20_50_NegotiatingAgent = lab_20_50.loc[lab_20_50['agentType'] == 'NegotiatingAgent']
lab_60_40_NegotiatingAgent = lab_60_40.loc[lab_60_40['agentType'] == 'NegotiatingAgent']
lab_40_60_NegotiatingAgent = lab_40_60.loc[lab_40_60['agentType'] == 'NegotiatingAgent']
lab_100_100_NegotiatingAgent = lab_100_100.loc[lab_100_100['agentType'] == 'NegotiatingAgent']

lab_10_10_RandomAgent = lab_10_10.loc[lab_10_10['agentType'] == 'RandomAgent']
lab_50_20_RandomAgent = lab_50_20.loc[lab_50_20['agentType'] == 'RandomAgent']
lab_20_50_RandomAgent = lab_20_50.loc[lab_20_50['agentType'] == 'RandomAgent']
lab_60_40_RandomAgent = lab_60_40.loc[lab_60_40['agentType'] == 'RandomAgent']
lab_40_60_RandomAgent = lab_40_60.loc[lab_40_60['agentType'] == 'RandomAgent']
lab_100_100_RandomAgent = lab_100_100.loc[lab_100_100['agentType'] == 'RandomAgent']

lab_10_10_SwarmAgent = lab_10_10.loc[lab_10_10['agentType'] == 'SwarmAgent']
lab_50_20_SwarmAgent = lab_50_20.loc[lab_50_20['agentType'] == 'SwarmAgent']
lab_20_50_SwarmAgent = lab_20_50.loc[lab_20_50['agentType'] == 'SwarmAgent']
lab_60_40_SwarmAgent = lab_60_40.loc[lab_60_40['agentType'] == 'SwarmAgent']
lab_40_60_SwarmAgent = lab_40_60.loc[lab_40_60['agentType'] == 'SwarmAgent']
lab_100_100_SwarmAgent = lab_100_100.loc[lab_100_100['agentType'] == 'SwarmAgent']
```

```python
average_10_10 = lab_10_10.groupby(['agentType'], as_index=True).mean()
ax = average_10_10['averageSpeed'].plot(kind='bar')
for p in ax.patches:
    ax.annotate(str(p.get_height().round(2)), (p.get_x(), p.get_height() + 0.02))
```

```python
average_50_20 = lab_50_20.groupby(['agentType'], as_index=True).mean()
ax = (average_50_20.drop(['RandomAgent']))['averageSpeed'].plot(kind='bar')
for p in ax.patches:
    ax.annotate(str(p.get_height().round(2)), (p.get_x(), p.get_height() + 0.02))
```
